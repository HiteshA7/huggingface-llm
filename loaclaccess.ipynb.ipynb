```python
from google.colab import userdata
hfl = userdata.get('deepseek')
```

```python
import os
os.environ["HF_TOKEN"] = hfl
os.environ["HUGGINGFACEHUB_API_TOKEN"] = hfl
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("deepseek-ai/deepseek-coder-1.3b-instruct")
model = AutoModelForCausalLM.from_pretrained("deepseek-ai/deepseek-coder-1.3b-instruct")
```

```python
from transformers import pipeline

pipeline = pipeline(task = "text-generation", model= model, tokenizer=tokenizer,device = 0)
```

```python
from langchain_huggingface import HuggingFacePipeline

HFPipeline = HuggingFacePipeline(pipeline = pipeline)
```

```python
prompt = """### Instruction:
Tell me which computer lanhuahe is the best to studfy in 2026 so that peopele could start studying that course and get skilledf up give some detailed explanation.

### Response:
"""

HFPipeline.invoke(prompt)
```

```python
from langchain_huggingface import ChatHuggingFace

output =  ChatHuggingFace(llm = HFPipeline)
```

```python
response = output.invoke([
    ("system", "You are a computer programmer with 10 yrs exp."),
    ("human", "Tell me which computer language is the best to study in 2026 so that people could start studying that course and get skilled up. Give some detailed explanation.")
])

print(response.content)
```
