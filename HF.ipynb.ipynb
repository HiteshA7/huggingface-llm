```python
os.environ["HF_TOKEN"] = ds
```

```python
import langchain
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
```

```python
regular =  HuggingFaceEndpoint(repo_id = "deepseek-ai/DeepSeek-V3.2",temperature = 0.3,max_new_tokens = 1000, task = "conversational")
```

```python
chat =  ChatHuggingFace(llm = regular)
```

```python
msg = chat.invoke([("system","you are a 10 yr experienced ai devoloper"),("user"),("Tell me why dont we directly use ChatHuggingFace while using it in langchain, instead we call the regular HuggingFaceEndpoint and then wrap it with the chatHuggingFace?")])
```

```python
msg.content
```
